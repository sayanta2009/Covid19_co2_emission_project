```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_energy_model = pd.read_excel('../../Dataset/processed_data/DataEnergyEmissions.xlsx')
data_energy_model.head()
```

```python
data_energy_model.info()
```

```python
X = data_energy_model.to_numpy()
X.shape
```

```python
plt.figure(figsize=(8, 4))
sns.distplot(data_energy_model['Energy_CO2_Emissions'])
plt.xlabel('Energy CO2 Emissions (Gg)')
plt.savefig('plots/energy_model/energy_co2_dist_plot.pdf')
plt.figure(figsize=(10, 4))
sns.lineplot(x='Year', y='Energy_CO2_Emissions', hue='Country', data=data_energy_model)
plt.ylabel('Energy CO2 Emissions (Gg)')
plt.savefig('plots/energy_model/energy_co2_vs_year.pdf')
```
