```python
import pandas as pd
data_Land = pd.read_excel('../../Dataset/processed_data/DataLandUseEmissions.xlsx')
data_Land.head()
```

```python
import matplotlib.pyplot as plt
from seaborn import scatterplot as scatter
import seaborn as sns
```

```python
g = sns.PairGrid(data_Land, hue="Year")
g = g.map(plt.scatter)
g = g.add_legend()
plt.savefig('plots/Land/pairplot.pdf')
```

```python
sns.distplot(data_Land ['Land_Use_Emissions'], color="g")
plt.savefig('plots/Land/total_distribution.pdf')
```

```python
sns.jointplot(x='Year',y='Land_Use_Emissions',data=data_Land,kind='hex', color="g")
plt.savefig('plots/Land/total_jointplot.pdf')
```

```python
# col_list = ['Land_Use_Emissions', 'Forestland', 'Harvested_Wood_Products', 'Cropland', 'Grassland',]
# data = pd.read_csv("DataLandUseEmissions.csv", usecols=col_list)
data_Land.corr()
```

```python
df = data_Land.drop(columns=['Year'])
```

```python
plt.figure(figsize=(18, 12))
sns.heatmap(df.corr(),annot=True)
plt.savefig('plots/Land/land_heatmap.pdf')
```

```python
plt.figure(figsize=(13, 7))
sns.lineplot(x='Year', y='Land_Use_Emissions', hue='Country', data=data_Land)
plt.savefig('plots/Land/total-Years.pdf')
```

```python
sns.distplot(data_Land['Forestland'], color ='g')
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Forestland', hue='Country', data=data_Land)
plt.savefig('plots/Land/Forestland_year.pdf')
```

```python
sns.distplot(data_Land['Harvested_Wood_Products'], color ='g')
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Harvested_Wood_Products', hue='Country', data=data_Land)
```

```python
sns.distplot(data_Land['Grassland'], color ='r')
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Grassland', hue='Country', data=data_Land)
```

```python
plt.figure(figsize=(13, 7))
sns.scatterplot(x="Forestland", y="Land_Use_Emissions", hue="Country", data=data_Land)
plt.savefig('plots/Industry/Forestland_Total.pdf')
```
