```python
import pandas as pd
data_industry = pd.read_excel('../../Dataset/processed_data/DataIndusProcessEmissions.xlsx')
data_industry.head()
```

```python
import matplotlib.pyplot as plt
from seaborn import scatterplot as scatter
import seaborn as sns
```

```python
g = sns.PairGrid(data_industry, hue="Year")
g = g.map(plt.scatter)
g = g.add_legend()
plt.savefig('plots/Industry/pairplot.pdf')
```

```python
sns.distplot(data_industry ['Industrial_Process_Emissions'], color="r")
plt.savefig('plots/Industry/total_distribution.pdf')
```

```python
sns.jointplot(x='Year',y='Industrial_Process_Emissions',data=data_industry,kind='hex', color="r")
plt.savefig('plots/Industry/total_jointplot.pdf')
```

```python
col_list = ['Industrial_Process_Emissions', 'Mineral_Industry', 'Chemical_Industry', 'Metal_Industry']
data = pd.read_excel('../../Dataset/processed_data/DataIndusProcessEmissions.xlsx', usecols=col_list)
data.corr()
```

```python
plt.figure(figsize=(18, 12))
sns.heatmap(data.corr(),annot=True)
plt.savefig('plots/Industry/industry_heatmap.pdf')
```

```python
plt.figure(figsize=(13, 7))
sns.lineplot(x='Year', y='Industrial_Process_Emissions', hue='Country', data=data_industry)
plt.savefig('plots/Industry/total-Years.pdf')
```

```python
sns.distplot(data_industry['Chemical_Industry'], color ='r')
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Chemical_Industry', hue='Country', data=data_industry)
plt.savefig('plots/Industry/Chemical_Industry_year.pdf')
```

```python
sns.distplot(data_industry['Mineral_Industry'], color ='r')
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Mineral_Industry', hue='Country', data=data_industry)
plt.savefig('plots/Industry/Mineral_Industry_year.pdf')
```

```python
sns.distplot(data_industry['Metal_Industry'], color ='r')
plt.figure(figsize=(12, 8))
sns.lineplot(x='Year', y='Metal_Industry', hue='Country', data=data_industry)
plt.savefig('plots/Industry/Metal_Industry_year.pdf')
```

```python
plt.figure(figsize=(13, 7))
sns.scatterplot(x="Metal_Industry", y="Industrial_Process_Emissions", hue="Country", data=data_industry)
plt.savefig('plots/Industry/Metal_Industry_Total.pdf')
```

```python
plt.figure(figsize=(16, 6))
sns.boxplot(data=data,palette='rainbow')
plt.savefig('plots/Industry/Industry_boxplot.pdf')
```
